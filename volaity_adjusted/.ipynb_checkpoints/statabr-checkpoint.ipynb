```python
"""
Implementation of Statistical Arbitrage in a trend-following way
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data
```

```python
# we use the 7 major currency pairs in the past five years
# we want to trade CAD-USD
trading_pairs = 'CADUSD=X'
symbols = ['AUDUSD=X', 'GBPUSD=X', 'CADUSD=X', 'CHFUSD=X','EURUSD=X', 'JPYUSD=X', 'NZDUSD=X']
start_date = '2015-01-01'
end_date = '2020-01-01'
```

```python
# load data for each currency
symbols_data = {}
for symbol in symbols:
    name = symbol+"_data.pkl"
    try:
        data1 = pd.read_pickle(name)
    except FileNotFoundError:
        data1 = data.DataReader(symbol,'yahoo',start_date,end_date)
        data1.to_pickle(name)
    symbols_data[symbol] = data1
```

```python
symbols_data
```

```python
# we visualize the graph
from itertools import cycle
plt.rcParams["figure.figsize"] = (20,10) # we set up the size

color = cycle('bgrcmky')
price_data = pd.DataFrame()

for symbol in symbols:
    mul = 1.0
    label = symbol+' ClosePrice'
    # since the values of usd to yen are too small
    # we multiply it by 100
    if symbol == 'JPYUSD=X':
        mul = 100.0
    price_data = price_data.assign(label = pd.Series(symbols_data[symbol]['Close'] * mul, index = symbols_data[symbol].index))
    ax = price_data['label'].plot(lw = 1.,color = next(color),label = label)
plt.legend()
```

```python
# we can see that the currencies are all correlatd to each other
# but in differet degrees, for example:
# the pairs CAD-USD, AUD-USD, NZD-USD are most correlated

# now, we want to see the correlation, simple moving average of each pair, and the corresponding price deviation
import statistics as stats
periods = 20 # we use 20-day SMA
price_history = {} # we use dictionary because we have 7 pairs

num_price_dev = [] # look back period of closeprice deviation
price_deviation = {}

num_days = len(symbols_data[trading_pairs].index)
correlation = {}

delta_projected_history = {} 
# it is the difference between projected closeprice and the real price per currency

final_delta = [] 
# the difference between projected closeprice of trading_pairs and the real price 
```

```python
# we define some parameters for our strategy
# +1: buy, -1: sell, 0: no action
orders = [] 
# positive: long, negative: short, 0: neutral
positions = [] 
# sum of total pnls, both closed and open
pnls = []
open_pnls = 0
closed_pnls = 0

# we define buy and sell thresholds
buy_entry = 0.01
sell_entry = -0.01
min_price_move = 0.01
num_shares = 100000 # the number of shares each order
min_profit_to_close = 10 # the close value for open pnl
last_sell_price = 0
last_buy_price = 0
position = 0
buy_sum_qty = 0
buy_sum_price_qty = 0
sell_sum_qty = 0
sell_sum_price_qty = 0
```

```python
# we calculate sma first
for i in range(num_days):
    close_prices = {}
    
    for symbol in symbols: # we loop each pair
        close_prices[symbol] = symbols_data[symbol]['Close'].iloc[i]
        if symbol not in price_history.keys():
            price_history[symbol] = []
            price_deviation[symbol] = []
        
        price_history[symbol].append(close_prices[symbol])
        # we check for sma periods
        if len(price_history[symbol])>periods:
            del(price_history[symbol][0])
        sma = stats.mean(price_history[symbol])
        price_deviation[symbol].append(close_prices[symbol]-sma)
        # since our look back period is 200 days, we check for the length of price deviation
        if len(price_deviation[symbol]) > num_days:
            del(price_deviation[symbol][0])
        
    # now, we want to check the relationship between our own trading pairs and other currency pairs
    # by using covariance and correlationwice
    # we loop t

    # covariance: shows how much two variables differ
    
    projcted_dev_from_sma = {}
    for symbol in symbols:
        if symbol == trading_pairs: # we don't need the correlation to itself
            continue
        correlation_label = trading_pairs + '<-' + symbol
        if correlation_label not in correlation.keys():
            correlation[correlation_label] = []
            delta_projected_history[correlation_label] = []

        # we check the length of projcted_dev_from_sma
        # we need at least two observations to compute correlation and covariance
        if len(price_deviation[symbol]) < 2:
            correlation[correlation_label].append(0)
            delta_projected_history[correlation_label].append(0)
            continue
        # we calculate correlation and covariance of the price deviation of two trading pairs
        corr = np.corrcoef(price_deviation[trading_pairs],price_deviation[symbol])
        cov_got = pd.Series(price_deviation[trading_pairs]).cov(pd.Series(price_deviation[symbol])) # we 
        # get the covariance between two series in their price deviation from sma
        corr_got = corr[0][1] # the correlation we get between two series
        
        # we append correlation
        correlation[correlation_label].append(corr_got)
        
        # we calculate projected price deviation for finding out delta
        # the projected price deviation will be calculated by cov * price_deviation
        projected_dev = cov_got * price_deviation[symbol][-1] # -1: the latest price deviation
        
        # delta: an estimate of the value of security changes with one dollar changes in its underlying assets
        # here we calculate projected delta: the projected price change - the actually price change
        
        # positive projected delta: price should have moved up more than it did
        # negative projected delta: price should have moved down more than it did
        delta_projected = projected_dev - price_deviation[symbol][-1] 
        delta_projected_history[correlation_label].append(delta_projected)
    
    # now we calculate the weight of each correlation pair
    sum_weights = 0 # we use the total weights for the use of normalization

    for symbol in symbols:
        # we skip the targeted trading pair still
        if symbol == trading_pairs:
            continue

        # we keep track of correlation
        correlation_label = trading_pairs + '<-' + symbol
        # why do we use correlation[correlation_label][-1]:
        # we use the latest correlation

        # why absolute numbers: some correlations are negative, but we only look at the weights
        # not whether it is positive or negative
        sum_weights += abs(correlation[correlation_label][-1])

    # we keep track of the total delta for weighing purposes
    final_delta_projected = 0
    close_price = close_prices[symbol]
    # we calculate the final delta now
    for symbol in symbols:
        if symbol == trading_pairs:
            continue

        correlation_label = trading_pairs + '<-' + symbol
        
        # now we weigh the final total deltas by using correlation
        # we multiply the delta_projected by the correlation value,
        # because we want to use it to weight the total delta
        
        # it is similar to we are long/short a whole bundle consisting of 6 pairs of trading instruments
        final_delta_projected += abs(correlation[correlation_label][-1]) * delta_projected_history[correlation_label][-1]
        
    # we normalize the final delta by the sum of correlation
    if sum_weights != 0:
        final_delta_projected /= sum_weights
    else:
        final_delta_projected = 0
    final_delta.append(final_delta_projected)
        
    # we begin our execution
    # if final delta < execution_sell_signal, we sell it
    # because we believe that the delta will go back to the one we have calculated 
    # the price should move down more, follwing our projection 
    # so we sell it
    # at that moment, stats arbitrage is similar to a trend-following strategy
    
    if (final_delta_projected < sell_entry and abs(close_price - last_sell_price) > min_price_move) or (position > 0 and (open_pnls > min_profit_to_close)):
        orders.append(-1)
        last_sell_price = close_price
        position -= num_shares
        sell_sum_qty += num_shares
        sell_sum_price_qty += num_shares*close_price
        print("Sell ", num_shares, " @ ", close_price,
        "Position: ", position)
        print("OpenPnL: ", open_pnls, " ClosedPnL: ", closed_pnls, "TotalPnL: ", (open_pnls + closed_pnls))

    # if final delta > execution_buy_signal, we buy it
    # because we believe that the delta will go back to the one we have calculated
    # the price should move up more
    # so we buy it

    elif (final_delta_projected > buy_entry and abs(close_price - last_buy_price) > min_price_move) or (position < 0 and (open_pnls > min_profit_to_close)):
        orders.append(1)
        last_buy_price = close_price
        position += num_shares
        buy_sum_qty += num_shares
        buy_sum_price_qty += num_shares*close_price
        print("Buy ", num_shares, " @ ", close_price,
        "Position: ", position)
        print("OpenPnL: ", open_pnls, " ClosedPnL: ", closed_pnls, "TotalPnL: ", (open_pnls + closed_pnls))
    else:
        orders.append(0)

    positions.append(position)

    # now we deal with pnls
    open_pnls = 0 # we update our open_pnl
    if position > 0: # we are holding long positions
        if sell_sum_qty > 0: # we have some sell trades against it 
            open_pnls = sell_sum_qty * (sell_sum_price_qty / sell_sum_qty - buy_sum_price_qty / buy_sum_qty)
        # we calculate open_pnls with respect to current close_price
        # that is, we close our position
        open_pnls += buy_sum_qty * (close_price - buy_sum_price_qty / buy_sum_qty)
    elif position < 0:
        if buy_sum_qty > 0: # we have some buy trades against it 
            open_pnls = buy_sum_qty * (sell_sum_price_qty / sell_sum_qty - buy_sum_price_qty / buy_sum_qty)
        open_pnls += sell_sum_qty * (sell_sum_price_qty / sell_sum_qty - close_price)
    else: # flat position, we just update our close pnls
        closed_pnls += sell_sum_price_qty - buy_sum_price_qty

        # we make them to 0 for a new buy-sell cycle
        buy_sum_price_qty = 0
        buy_sum_qty = 0
        sell_sum_price_qty = 0
        sell_sum_qty = 0
        last_buy_price = 0
        last_sell_price = 0
    pnls.append(closed_pnls+open_pnls)
            
```

```python
# now we visualize the wieghted graph 
correlation_data = pd.DataFrame()

for symbol in symbols:
    if symbol == trading_pairs:
        continue
    correlation_label = trading_pairs + '<-' + symbol

    correlation_data = correlation_data.assign(label = pd.Series(correlation[correlation_label], index = symbols_data[symbol].index))
    # we print the correlation relationshipout
    # since we don't set the assign to be inplace, we need to plot the data every time
    correlation_data['label'].plot(color = next(color),lw = 1, label = "Correlation "+correlation_label)
plt.legend()

# in order ot make it more clear, we draw some lines
for i in np.linspace(-1,1,num = 8): # we use np.linspace for loop through a series of equal intervals with length of 0.25
    plt.axhline(i,lw = 0.5,color = 'k')
```

```python
# we can see that AUD-USD and NZD-USD are two most correlated pairs
# how about delta?
# we visualize it
delta_actual = pd.DataFrame()
for symbol in symbols:
    if symbol == trading_pairs:
        continue
    
    projection_label = trading_pairs + '<-' + symbol
    delta_actual = delta_actual.assign(signal = pd.Series(delta_projected_history[projection_label],index = symbols_data[symbol].index))
    delta_actual['signal'].plot(color = next(color),lw = 1,label = "delta projected: "+projection_label)
plt.legend()

# from observing the following graph,
# we can see that the deviations are mainly from GBPUSD and CHFUSD
# but the predictions may not be good, as we have seen from above
# they don't have a good correlation with our trading pair
# we then check our positions, trades, signal, PnL
```

```python
delta_actual = delta_actual.assign(close = pd.Series(symbols_data[trading_pairs].Close))
delta_actual = delta_actual.assign(order = pd.Series(orders,index = symbols_data[symbol].index))
delta_actual = delta_actual.assign(signal = pd.Series(final_delta,index = symbols_data[symbol].index))
delta_actual = delta_actual.assign(positions = pd.Series(positions,index = symbols_data[symbol].index))
delta_actual = delta_actual.assign(PnL = pd.Series(pnls,index = symbols_data[symbol].index))
delta_actual
```

```python
# now we plot our trades
# note: the close price here is the difference between our trading pairs and the sma
plt.plot(delta_actual.close, color = 'k',lw=1,label = 'price')
plt.plot(delta_actual.close[delta_actual.order==1],color = 'r',marker = '^',markersize = 7,lw = 0,label = 'buy')
plt.plot(delta_actual.close[delta_actual.order==-1],color = 'g',marker = 'v',markersize = 7,lw = 0,label = 'sell')
plt.legend()
```

```python
# from that graph we cannot get more information
# so we draw the graph for our signals
plt.plot(delta_actual.signal,color = 'k',label = 'signal')
plt.plot(delta_actual.signal[delta_actual.order==1],color = 'r',marker = '^',markersize = 7,lw = 0,label = 'buy')
plt.plot(delta_actual.signal[delta_actual.order==-1],color = 'g',marker = 'v',markersize = 7,lw = 0,label = 'sell')
plt.axhline(y = 0,color = 'k')

# we draw some lines about buy / sell entry 
for i in np.arange(buy_entry,buy_entry*10,buy_entry*2):
    plt.axhline(y = i,color = 'r',lw = 0.5) # we are doing a trend following strategy
    
for i in np.arange(sell_entry,sell_entry*10,sell_entry*2):
    plt.axhline(y = i,color = 'g',lw = 0.5) # we are doing a trend following strategy
    
# we can see most buy/sell behaviors are still in their own entry area
# there are also some mixes, because we have reached the min profits to close
```

```python
# let's check out position
plt.plot(delta_actual.positions, color = 'k',lw=1,label = 'positions')
plt.plot(delta_actual.positions[delta_actual.positions > 0],color = 'r',marker = '^',markersize = 7,lw = 0,label = 'long')
plt.plot(delta_actual.positions[delta_actual.positions < 0],color = 'g',marker = 'v',markersize = 7,lw = 0,label = 'short')
plt.plot(delta_actual.positions[delta_actual.positions == 0],color = 'k',marker = '.',markersize = 7,lw = 0,label = 'flat')
plt.legend()

# we draw some lines about long / short positions
plt.axhline(y = 0,color = 'k',lw = 0.5)
for i in np.arange(num_shares,num_shares*5,num_shares):
    plt.axhline(y = i,color = 'r',lw = 0.5) # we are doing a trend following strategy
    
for i in np.arange(-num_shares,-num_shares*5,-num_shares):
    plt.axhline(y = i,color = 'g',lw = 0.5) # we are doing a trend following strategy
```

```python
# then we check our PnL graph
plt.plot(delta_actual.PnL, color = 'k',lw=1,label = 'PnL')
plt.plot(delta_actual.PnL[delta_actual.PnL > 0],color = 'g',marker = '.',markersize = 7,lw = 0,label = 'profits')
plt.plot(delta_actual.PnL[delta_actual.PnL < 0],color = 'r',marker = '.',markersize = 7,lw = 0,label = 'loss')
plt.legend()
```
