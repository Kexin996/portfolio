```python
"""
Measure of Risk in Algo Trading by Checking the Mean-Reversion Volatility-Adjusted Strategy
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# we first check the export csv file from our mean-reversion volatility-adjusted strategy first
res = pd.read_csv('amzn_data_risk.csv')
res
```

```python
# 1. stop-loss / max-loss
# the first risk we gonna check is stop loss
# the maximum loss our strategy can have during a time period
# the time of stop depends on the time frame we choose
# if we set the stop loss for the day, the algorithm gonna resume trading the next trading day
# same logic for weeks, months...

# we now compute the weekly and monthly losses.
days = len(res.index)
pnl = res['PnL']

weekly_loss = []
monthly_loss = []
# since there are only five trading days in a week and twenty trading days in a month,
# we calculate the weekly loss and monthly loss as in the following:
for i in range(days):
    if i >= 5 and pnl[i-5] > pnl[i]: 
        weekly_loss.append(pnl[i]-pnl[i-5]) 
        # note: it is not the actual 'weekly loss' or 'monthly loss'
        # it is a rolling loss ---> the maximum loss we can have
        # in a 5-day or 20-day time frame
    if i >= 20 and pnl[i-20] > pnl[i]: 
        monthly_loss.append(pnl[i]-pnl[i-20])

plt.rcParams["figure.figsize"] = (20,10) # we set up the runtime configuration parameters
# we visualize them
plt.hist(weekly_loss,50) # we set 50 as the number of bins in order to see the frequencly more clearly
plt.gca().set(title='Weekly Loss Distribution', xlabel='$',ylabel='Frequency')
plt.show()

plt.hist(monthly_loss,50)
plt.gca().set(title = 'Monthly Loss Distribution',xlabel = '$',ylabel = 'Frequency')

# from the graphs, we can see:
# 1. for weekly loss, it is normal to lose smaller than $100
# 2. for monthly loss, it is normal to lose smaller than $150
# for other amount of money, they are unlikely
# when this kind of stop loss happens, we need to be careful
```

```python
# 2. Max drawdown
# the maxmimum loss the strategy can take between a trough and a peak (highest pnl - lowest pnl during a period)
max_pnl = 0
max_drawdown = 0
# we also record the pnl values at max_drawdown
drawdown_high = 0
drawdown_low = 0

for i in range(days):
    # we decide whether the pnl in the loop is larger than our current max_pnl
    max_pnl = max(max_pnl,pnl[i])
    # since we are recording maximum loss, we use the value of (max_pnl - curren_pnl)
    drawdown = max_pnl - pnl[i]
    # we check whether we have updated our max drawndown
    if drawdown > max_drawdown:
        max_drawdown = drawdown
        drawdown_high = max_pnl
        drawdown_low = pnl[i]

# we visualize it
res['PnL'].plot(lw = 1)
plt.axhline(drawdown_high,color = 'g')
plt.axhline(drawdown_low,color = 'r')
print("Maximum drawdown:",max_drawdown)
```
