```python
"""
Measure of Risk in Algo Trading by Checking the Mean-Reversion Volatility-Adjusted Strategy
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# we first check the export csv file from our mean-reversion volatility-adjusted strategy first
res = pd.read_csv('amzn_data_risk.csv')
res
```

```python
# 1. stop-loss / max-loss
# the first risk we gonna check is stop loss
# the maximum loss our strategy can have during a time period
# the time of stop depends on the time frame we choose
# if we set the stop loss for the day, the algorithm gonna resume trading the next trading day
# same logic for weeks, months...

# we now compute the weekly and monthly losses.
days = len(res.index)
pnl = res['PnL']

weekly_loss = []
monthly_loss = []
# since there are only five trading days in a week and twenty trading days in a month,
# we calculate the weekly loss and monthly loss as in the following:
for i in range(days):
    if i >= 5 and pnl[i-5] > pnl[i]: 
        weekly_loss.append(pnl[i]-pnl[i-5]) 
        # note: it is not the actual 'weekly loss' or 'monthly loss'
        # it is a rolling loss ---> the maximum loss we can have
        # in a 5-day or 20-day time frame
    if i >= 20 and pnl[i-20] > pnl[i]: 
        monthly_loss.append(pnl[i]-pnl[i-20])

plt.rcParams["figure.figsize"] = (20,10) # we set up the runtime configuration parameters
# we visualize them
plt.hist(weekly_loss,50) # we set 50 as the number of bins in order to see the frequencly more clearly
plt.gca().set(title='Weekly Loss Distribution', xlabel='$',ylabel='Frequency')
plt.show()

plt.hist(monthly_loss,50)
plt.gca().set(title = 'Monthly Loss Distribution',xlabel = '$',ylabel = 'Frequency')

# from the graphs, we can see:
# 1. for weekly loss, it is normal to lose smaller than $100
# 2. for monthly loss, it is normal to lose smaller than $150
# for other amount of money, they are unlikely
# when this kind of stop loss happens, we need to be careful

print('Maximum weekly loss:',min(weekly_loss))
print('Maximum month loss:',min(monthly_loss))
```

```python
# 2. Max drawdown
# the maxmimum loss the strategy can take between a trough and a peak (highest pnl - lowest pnl during a period)
max_pnl = 0
max_drawdown = 0
# we also record the pnl values at max_drawdown
drawdown_high = 0
drawdown_low = 0

for i in range(days):
    # we decide whether the pnl in the loop is larger than our current max_pnl
    max_pnl = max(max_pnl,pnl[i])
    # since we are recording maximum loss, we use the value of (max_pnl - curren_pnl)
    drawdown = max_pnl - pnl[i]
    # we check whether we have updated our max drawndown
    if drawdown > max_drawdown:
        max_drawdown = drawdown
        drawdown_high = max_pnl
        drawdown_low = pnl[i]

# we visualize it
res['PnL'].plot(lw = 1)
plt.axhline(drawdown_high,color = 'g')
plt.axhline(drawdown_low,color = 'r')
print("Maximum drawdown:",max_drawdown)
```

```python
# 3. Positions Limits
# Maximum Long / Short Position
# we quantify the maximum our long / short positions can be based on historical data
# we visualize our positions
res['Position'].hist(bins=20)
plt.gca().set(title='Position Distribution', xlabel='Shares',ylabel='Frequency')
print('Maximum positions:',abs(min(res['Position'])))
```

```python
# from the graph above, we can see that most of the time we don't hold a position
# and for short positions, the maximum is 60 ---> if the position is more than it, we need to be cautious
# for long positions, if it exists, we need to be cautious (but our strategy doesn't profit at all hahaha)
```

```python
# 4. Position holding time
# it is the length of time between holding a position and closing it or even turning it into oppose positions
# we record the length of holding positions and visualize them
positions = res['Position']
current_pos = 0
current_pos_start = 0 # record the start and end date
holding = []
for i in range(days):
    pos = positions[i]
    
    # we first check current_pos
    # if it is 0 (flat), we need to begin a new interval
    if current_pos == 0:
        # unless pos is not 0
        if pos != 0:
            current_pos = pos
            current_pos_start = i
    else:
        # we check whether the positions change its sign
        # including 0 (e.g. from long positions to flat)
        if pos*current_pos <= 0:
            # we append the interval
            holding.append(i-current_pos_start)
            # we making our current positions to pos 
            current_pos = pos
            # and we update the beginning of the new interval
            current_pos_start = i

# now we visualize the positions and its frequency
plt.hist(holding,50)
plt.gca().set(title = 'Position Holding Time Distribution', xlabel='Holdingmtime days', ylabel='Frequency')
print('Maximum Position Holding Time:',max(holding))
    
```

```python
# from the graph, we can see the smallest holding days is around 100
# and the largest days is around 250
# if the amount of position holding time exceeds this limit, we need to be cautious
```

```python
# 5. Variance of PnLs
# the standard deviation of returns in different timeframes, e.g. days, weeks, months, etc.
# now we calculate a weekly std of returns
weekly_pnls = []
last_week = 0 # we initialize the variables
# this variable is used to set the time frame (a week)

for d in range(days):
    if d-last_week == 5: # a week
        weekly_pnls.append(pnl[d]-pnl[last_week])
        # we update the last_week
        last_week = d

# we calculate the standard deviation
from statistics import stdev,mean
print("Weekly PnLs Standard Deviation: ",stdev(weekly_pnls))
# we plot the weekly_pnls
plt.hist(weekly_pnls,bins = 20)
plt.title('Frequency of Weekly PnLs')
# from the graph we can see, most of the time the strategy is losing money.
# the ditribution is skewed to the left ----> loss
# this confirms the results from this strategy
```

```python
# 6. Sharpe Ratio / Sortino Ratio
# sharpe_ratio = average_daily_PnLs / std_daily_PnLs
# it measures the profitability (average) of the strategy with the total volatility (std)
# it is for low-volatility portfolio

# sortino_ratio is very similar to sharpe ratio, but:
# sortino_ratio = average_daily_PnLs / std_daily_loss
# it doesn't consider positive volatility,
# because risks only come from loss
# only downside movements are consideration in the dominator
# it is for high-volatility portfolio


# in our calculation, we calculate the weekly sharpe ratio and sortino ratio
last_week = 0
weekly_pnls = []
weekly_loss = []
for d in range(days):
    if d - last_week == 5:
        weekly_pnls.append(pnl[d]-pnl[last_week])
        # we check whether it is a loss or not
        if pnl[d]-pnl[last_week] < 0:
            weekly_loss.append(pnl[d]-pnl[last_week])
        last_week = d

# we calculate the sharpe ratio and sortino ratio
sharpe_ratio = mean(weekly_pnls) / stdev(weekly_pnls)
print('The sharpe ratio is:',sharpe_ratio)
sortino_ratio = mean(weekly_pnls) / stdev(weekly_loss)
print('The sortino ratio is:',sortino_ratio)
# we can see that the sharpe ratio and sortino ratio are almost the same
# menaing that the volatility of the strategy mainly comes from negative volatility
# the strategy only has an extremely small portion of profits
```

```python
# 7. Maximum executions per period
# we use it to prevent over trading
last_week = 0
executions = []
executions_this_week = 0
for i in range(days):
    if i-last_week < 5: # we record the number of executions, not orders
        if res['Orders'][i] > 0:
            executions_this_week += 1
    else:
        executions.append(executions_this_week)
        executions_this_week = 0
        last_week = i
plt.hist(executions,10)
# we can see that most of the week time frames, the strategy doesn't act at all
# the maximum execution in a week is 4
plt.show()

# how about the monthly executions?

last_month = 0
executions = []
executions_this_month = 0
for i in range(days):
    if i-last_month < 20: # we record the number of executions, not orders
        if res['Orders'][i] > 0:
            executions_this_month += 1
    else:
        executions.append(executions_this_month)
        executions_this_month = 0
        last_month = i
plt.hist(executions,10)
# from the graph, we can see the maximum executaions in a month is 6
# however, from these two graphs, we can see our strategy is not active at all
# thus, we need to be cautious for suddent active execuations
```

```python
# 8. Volume Limits
# the total number of positions traded during the whole strategy
# it is used to prevent over trading
traded_vol = 0
for d in range(days):
    # we check for the volume when we execute an order
    if res['Orders'][d] != 0:
        traded_vol += abs(res['Position'][d] - res['Position'][d-1])
print('Total traded volume:',traded_vol)

# in total we just trade a volume of 320...
# we are under-trading....
# it doesn't not exceed the amount we expect from maximum executions we calculated before
```
