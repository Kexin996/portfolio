```python
"""
Implementation of mean-reversion strategy with absolute price oscillator while dynamically adjusting for volatility
(Dynamically adjust for risks)
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats
import math as math
from pandas_datareader import data
```

```python
# we load the dataset 
# we use the same dataset for comparison
try:
    data_amzn = pd.read_pickle('amzn_data.pkl')
except:
    data_amzn = data.DataReader("AMZN",'yahoo','2015-01-01','2020-01-01')
    data_amzn.to_pickle('amzn_data.pkl')
data_amzn
```

```python
# before the preparation, we just observe the dataset first
plt.rcParams["figure.figsize"] = (20,10) # we set up the runtime configuration parameters
data_amzn['Close'].plot(color = 'k',lw = 1)
```

```python
# then we check the total standard deviation
data_amzn['Close'].rolling(window = 20, min_periods = 1).std().plot(color = 'b',lw = 1)
total_mean = np.mean(data_amzn['Close'].rolling(window = 20, min_periods = 1).std())
plt.axhline(total_mean,lw = 1)
total_mean

# for convenience, even the mean_std = 1.5388011500701388, we will use 1.54
# 1.54 will be the standard we use to assess the volatility
```

```python
# this time, we need to define low / high risks limits, and the increments
min_num_shares = 1 # the minimum share we allow for a trade, e.g. for a bad month
max_num_shares = 50
increment_shares = 2
num_shares = min_num_shares # we begin with very small values
num_shares_history = []
abs_position = [] # we use it to record the absolute value of our positions
# throughout the trade

# we also set up the increment for each risk limit. in the following
risk_limit_weekly = -95 
increment_risk_limit_weekly = -189 # for example, in good month, the weekly loss can be extended to -189-95 = =284
risk_limit_monthly = -161
increment_risk_limit_monthly = -322

# we set up the max_position to be 5 
risk_limit_max_position = 5
# and we gradually add 2 to it
increment_risk_limit_max_position = 3
max_position = []

# for risk_limit_shares, we repeat the same process
risk_limit_shares = 5
increment_risk_limit_shares = 2
max_shares_size = [] # we use it to keep track of the change in the max limits of our shares at each trade

# we don't use the risk_limit_trading_volume
# since now our strategy is dynamic in positions
# the risk_limit_trading_volume is not that useful, as our strategy is dynamic
# but we still keep the max_holding_time
risk_limit_max_holding_time = 221* 3 


# we also use some variables as 'flags' to check the risk condition
risk_violated = False # default value

traded_volume = 0
current_pos = 0
current_pos_start = 0

last_risk_change_index = 0 # we use it to dynamically adjust the risk management

# in the following of the code, we will check each limit step by step
```

```python
# for the preparation, it is the same as the strategy without adjustment
# define the window_size for fast_ema and slow_ema
# for fast_ema, it is 10 days
fast = 10
k_fast = 2 / (fast+1)
ema_fast = 0
ema_fast_values = []

slow = 40 # 40 days for slow_ema
k_slow = 2 / (slow+1)
ema_slow = 0
ema_slow_values = []
apo_values = [] # we save it for buying / selling signals
```

```python
# we save some variables for storing orders, positions, and profit & loss
orders = [] 
# -1: sell order
# 1: buy order
# 0: neutral

positions = []
# -1: short
# 1: long
# 0: netural

# open value - closed value
# sum of open pnl and closed pnl (lock-in profits - realized)
pnls = []

# we record the last buy/sell price to prevent overtrading
# we don't want to buy/sell just around the last buy/sell trade to ensure profits
last_buy_price = 0
last_sell_price = 0

# current position
position = 0

# summation quantity since last no position
buy_sum_qty = 0
sell_sum_qty = 0

buy_sum_price_qty = 0
sell_sum_price_qty = 0

# record open and closed pnls
open_pnls = 0
closed_pnls = 0
```

```python
# we set up threshold for enter positions and the minimum price change since the last trade
apo_buy_entry = -5
apo_sell_entry = 5
min_price_movement = 5 # we use it prevent overtrading
min_profit_to_close = 5*num_shares 

```

```python
# now, we slighly modify the way we calculate ema
# we use std /  * k_fast / k_slow as the real factor
# we need a simple moving average for updating the standard deviation
# and we define the periods to be 20 days
sma_periods = 20
price_history = []
```

```python
close = data_amzn['Close']
for close_price in close:
    price_history.append(close_price)
    if len(price_history) > sma_periods:
        # we delete the first element to update the std
        del(price_history[0])
    
    # we calculate the variance
    sma = stats.mean(price_history)
    variance = 0
    for temp_price in price_history:
        variance += (temp_price - sma)**2
    
    std = (variance / len(price_history)) ** (0.5)
    # here, we define the std_factor as:
    # std_factor = std / 1.10
    std_factor = std / 1.538
    # if std_factor = 0, for our later calculation, we just set it to 1
    if std_factor == 0:
        std_factor = 1
    
    # we check whether the data is the first observation
    if ema_fast == 0:
        ema_fast = close_price
        ema_slow = close_price
    else:
        # we use our std_factor here
        ema_fast = k_fast*std_factor*close_price + (1-k_fast*std_factor)*ema_fast
        ema_slow = k_slow*std_factor*close_price + (1-k_slow*std_factor)*ema_slow
    
    ema_fast_values.append(ema_fast)
    ema_slow_values.append(ema_slow)
    
    # calculate apo
    apo = ema_fast - ema_slow
    apo_values.append(apo)
    
    # at this point, we check for the number of shares
    # apparently, this condition will not be triggered
    # since we set the risk_limit_shares to be 10 * 1.5, but num_shares to be 10
    # this condition in this simplified example is only for showing that we are checking 
    # risk_limit_shares
    if num_shares > risk_limit_shares:
        print("Risk Violation: risk_limit_shares")
        print(num_shares,"> risk_limit_shares",risk_limit_shares)
        risk_violated = True
    
    min_profit_to_close = 5*num_shares 
    # now, the min_profit_to_lose becomes a dynamically one
    # as we set num_shares to be dynamic

    # we consider std_factor in the conditions of our buy / sell actions
    # we only execute if we don't violate our risk limits
    if (not risk_violated and (apo > apo_sell_entry * std_factor and abs(close_price - last_sell_price) > min_price_movement * std_factor) \
    or (position > 0 and (apo >= 0 or open_pnls > min_profit_to_close / std_factor))):
        # why we multipy std_factor for the first condition: because now we are considering whether to enter a position
        # smaller std_factor ---> smaller volatility, meaning that we are likely to enter the position
        
        # why do we divide std_factor for the second condition:
        # we tend to leave with larger volatility
        orders.append(-1)
        last_sell_price = close_price
        
        # since our num_shares is now dynamic
        # we need to separate our positions in two conditions:
        # 1. we enter a short postition.
        # that means our current position is flat
        if position == 0:
            position -= num_shares
            sell_sum_qty += num_shares
            sell_sum_price_qty += num_shares*close_price
        
            # we updated our traded volumnes of we execute trade
            # for later we checking the maximum volumnes
            traded_volume += num_shares
            print("Sell ",num_shares, "@",close_price)
            print("Position:",position)
        else: # 2. we close out buy position
            # why do we separate two situations, one we enter short position, and when we close out long position?
            # because now the num_shares is dynamic
            # the positions we enter and exit are different
            
            sell_sum_qty += position # now the position is positive, since we denote long position as positive
            sell_sum_price_qty += position*close_price

        
            # we updated our traded volumnes of we execute trade
            # for later we checking the maximum volumnes
            traded_volume += position
            print("Sell ",position, "@",close_price)
            print("Position:",position)
            position = 0 # close out position, position becomes flat
            
    # we apply the same logic to buy order
    elif (not risk_violated and (apo < apo_buy_entry * std_factor and abs(close_price - last_buy_price) > min_price_movement * std_factor) \
    or (position < 0 and (apo <= 0 or open_pnls > min_profit_to_close / std_factor))):
        orders.append(1)
        last_buy_price = close_price
        if position == 0: # we enter a buy position
            position += num_shares
            buy_sum_qty += num_shares
            buy_sum_price_qty += num_shares*close_price
        
            # we updated our traded volumnes of we execute trade
            # for later we checking the maximum volumnes
            traded_volume += num_shares
        
            print("Buy ",num_shares, "@",close_price)
            print("Position:",position)
        else:
            # we close out short position
            buy_sum_qty += abs(position) 
            # note: for short position, it is negative, and for quantity, we need
            # to use absolute values
            buy_sum_price_qty += abs(position)*close_price
            traded_volume += abs(position)
            print("Buy ",abs(position), "@",close_price)
            print("Position:",abs(position))
    
            position = 0 # we turn the position to flat
            
    else: # flat position
        orders.append(0)
    positions.append(position)
    
    # after sending orders, we need to check the limits of maximum positions, maximum holding time, and traded volume
    if current_pos == 0:
        # if it is 0, we update it
        current_pos = position
        current_pos_start = len(positions) # the start of our positions
        # we use it for calculating the holding time
    
    # we check whether we change the sign of our positions
    # or we flat our positions
    # it will not influence the previous condition
    if current_pos * position <= 0:
        # we record the position and position holding time
        current_pos = position # we update the position, since we begin a new interval
        current_holding_time = len(positions) - current_pos_start
        current_pos_start = len(positions)
        
        
        # we check the maximum holding time
        if current_holding_time > risk_limit_max_holding_time:
            print("Risk Violation: risk_limit_max_holding_time")
            print(current_holding_time,"> risk_limit_max_holding_time",risk_limit_max_holding_time)
            risk_violated = True
    # now we check the maximum positions and traded volumnes
    if abs(position) > risk_limit_max_position: # note: we are checking the numbers, so we use positive values
        print("Risk Violation: risk_limit_max_position")
        print(abs(position),"> risk_limit_max_position",risk_limit_max_position)
        risk_violated = True
    

    
        
    
    # for open_pnls, we don't have to change it
    # but we need to check for the weekly and monthly limit
    
    open_pnls = 0
    
    if position > 0:
        # we need to minus some closed pnls if position > 0 but sell_qty > 0
        # because that means our position is long, but we have done some selling against this trade
        # we will minus the closed pnls
        if sell_sum_qty > 0:
            # we use an average price difference for calculation
            open_pnls += sell_sum_qty*(sell_sum_price_qty / sell_sum_qty - buy_sum_price_qty / buy_sum_qty)
        # then we set the remaining to open pnls with respect to market price
        
        # why do we use (sell_sum_qty - buy_sum_qty): we want the unrealized buy_sum_qty
        # why do we use (close_price - buy_sum_price / buy_sum_qty):
        # because we want to calculate the profits if we close at current price
        # we want to sell at close_price because we have long position
        open_pnls += abs(sell_sum_qty - position) * (close_price - buy_sum_price_qty / buy_sum_qty)
    
    # we now look at the case for short position
    elif position < 0:
        # that means we have some realized buy orders
        if buy_sum_qty > 0:
            open_pnls += buy_sum_qty * (sell_sum_price_qty / sell_sum_qty - buy_sum_price_qty / buy_sum_qty)
        
        # we want to calculate the unrealized buy open pnls
        
        # why do we use (sell_sum_price_qty / sell_sum_qty-open_price):
        # now out position is short, and we want to buy back at the current open_price 
        open_pnls += abs(buy_sum_qty - position) * (sell_sum_price_qty / sell_sum_qty-close_price)
    else:
        # the position is 0, we just update close_pnls
        # open_pnls is 0. we can directly use sell_sum_price_qty - buy_sum_price_qty
        closed_pnls += (sell_sum_price_qty - buy_sum_price_qty)
        
        # and we reset the sum_price_qty 
        buy_sum_price_qty = 0
        sell_sum_price_qty = 0
        
        buy_sum_qty = 0
        sell_sum_qty = 0
        
        last_buy_price = 0
        last_sell_price = 0
    print("OpenPnL: ", open_pnls, " ClosedPnL: ", closed_pnls)
    print("Totl PnL:",closed_pnls+open_pnls)
    pnls.append(closed_pnls+open_pnls)
    
    # we decide whether this month is a 'good' month to adjust the risk limits
    if len(pnls) >= 20: # we check whether we have a month of PnLs, that is, 20 days of PnLs
        monthly_pnls = pnls[-1]-pnls[-20] # gets the latest pnls of the month
        
        if len(pnls) - last_risk_change_index == 20: # we check whether we have a month since the last time
            # we change our risk measure
            if monthly_pnls > 0: # we check whether it is a good month. positive PnL means that we have a good month.
                # it is. we extend our risk limits
                num_shares += increment_shares
                # we check whether we have exceeded the maximum number of shares we have set
                # if not, we can extend
                if num_shares <= max_num_shares:
                    risk_limit_weekly += increment_risk_limit_weekly
                    risk_limit_monthly += increment_risk_limit_monthly
                    risk_limit_max_position += increment_risk_limit_max_position
                    risk_limit_shares += increment_risk_limit_shares
                else:
                    num_shares = max_num_shares # else, we don't extend and fix it in the max_num_shares
            elif monthly_pnls < 0: # check whether it is a bad month
                # it is. we squeeze our risk limits
                num_shares -= increment_shares
                # we check whether the num_shares is below the minimum number of shares we have set or not
                # if not, we can squeeze
                if num_shares >= min_num_shares:
                    risk_limit_weekly -= increment_risk_limit_weekly
                    risk_limit_monthly -= increment_risk_limit_monthly
                    risk_limit_max_position -= increment_risk_limit_max_position
                    risk_limit_shares -= increment_risk_limit_shares
                # else: we just fix it in the minimum
                else:
                    num_shares = min_num_shares
            # if it is neither a good month nor a bad month, we don't move our num_shares
            # we update the index, showing that we are going to a new interval of a trading month
            last_risk_change_index = len(pnls)
            
    # we record those values for later visualization
    num_shares_history.append(num_shares)
    abs_position.append(abs(position))
    max_position.append(risk_limit_max_position)
    max_shares_size.append(risk_limit_shares)
    
                    
    # we check for our weekly and monthly pnls
    # it doesn't conflict with the good/bad months checks we have written above
    if len(pnls) >= 5:
        weekly_pnls = pnls[-1]-pnls[-5]
        if weekly_pnls < risk_limit_weekly: # we use '<' because both values are negative
            print("Risk Violation: risk_limit_weekly")
            print(abs(weekly_pnls),"> risk_limit_weekly",abs(risk_limit_weekly))
            risk_violated = True
    
    if len(pnls) >= 20:
        monthly_pnls = pnls[-1]-pnls[-20]
        if monthly_pnls < risk_limit_monthly: # we use '<' because both values are negative
            print("Risk Violation: risk_limit_monthly")
            print(abs(monthly_pnls),"> risk_limit_monthly",abs(risk_limit_monthly))
            risk_violated = True
    
    # this part I will revisit later, as the book doesn't explain whether it is possible for
    # the strategy to restart as we meet the condition again after reaching risk-violation conditions
```

```python
# now we want to visualize the result
data_amzn['Fast_EMA'] = pd.Series(ema_fast_values,index = data_amzn.index)
data_amzn['Slow_EMA'] = pd.Series(ema_slow_values,index = data_amzn.index)
data_amzn['APO'] = pd.Series(apo_values,index = data_amzn.index)
data_amzn['Orders'] = pd.Series(orders,index = data_amzn.index)
data_amzn['Position'] = pd.Series(positions,index = data_amzn.index)
data_amzn['PnL'] = pd.Series(pnls,index = data_amzn.index)
data_amzn['MaxShares'] = pd.Series(max_shares_size, index = data_amzn.index)
data_amzn['NumShares'] = pd.Series(num_shares_history, index = data_amzn.index)
data_amzn['AbsPosition'] = pd.Series(abs_position, index = data_amzn.index)
data_amzn['MaxPosition'] = pd.Series(max_position, index = data_amzn.index)
```

```python
# we first check our new strategy's buy and sell
data_amzn['Close'].plot(color = 'k',lw = 1)
data_amzn['Fast_EMA'].plot(color = 'y',lw = 1)
data_amzn['Slow_EMA'].plot(color = 'm',lw = 1)
# we show our order
plt.plot(data_amzn.loc[data_amzn.Orders == 1].index, data_amzn.Close[data_amzn.Orders == 1],lw = 0,label = 'buy', markersize = 7,marker = '^',color = 'r')
plt.plot(data_amzn.loc[data_amzn.Orders == -1].index, data_amzn.Close[data_amzn.Orders == -1],lw = 0,label = 'sell', markersize = 7,marker = 'v',color = 'g')
plt.legend()
# it is almost the same
```

```python
# we also display the apo values and buy/sell orders
data_amzn['APO'].plot(color ='b',lw = 1)
plt.plot(data_amzn.loc[data_amzn.Orders == 1].index, data_amzn.APO[data_amzn.Orders == 1],color = 'r',marker = '^',markersize = 7,lw = 0, label = 'buy')
plt.plot(data_amzn.loc[data_amzn.Orders == -1].index, data_amzn.APO[data_amzn.Orders == -1],color = 'g',marker = 'v',markersize = 7,lw = 0, label = 'sell')

# we draw some horizontal lines to see where to buy and where to sell more clearly
plt.axhline(y = 0, lw = 0.5, color = 'k')

for i in range(apo_buy_entry, apo_buy_entry*5,apo_buy_entry):
    plt.axhline(y = i,lw = 0.5, color = 'r')

    
for i in range(apo_sell_entry, apo_sell_entry*5,apo_sell_entry):
    plt.axhline(y = i,lw = 0.5, color = 'g')
```

```python
# this time, we can see that althogh our buy/sell behaviors don't change
# the amount of positions have changed a lot
# that explains the better performance:
# in the bad months, we just trade less
data_amzn['Position'].plot(color = 'k',lw = 0.5)
plt.plot(data_amzn[data_amzn.Position == 0].index, data_amzn.Position[data_amzn.Position == 0],marker = '.',markersize = 4,color = 'k',lw = 0, label = 'flat')
plt.plot(data_amzn[data_amzn.Position > 0].index, data_amzn.Position[data_amzn.Position >0],marker = '^',markersize = 4,color = 'r',lw = 0, label = 'buy')
plt.plot(data_amzn[data_amzn.Position < 0].index, data_amzn.Position[data_amzn.Position <0],marker = 'v',markersize = 4,color = 'g',lw = 0, label = 'sell')

plt.axhline(y = 0, lw = 0.5, color = 'k')

for i in range(num_shares, num_shares*5,num_shares):
    plt.axhline(y = i,lw = 0.5, color = 'r')

for i in range(0, -num_shares*5,-num_shares):
    plt.axhline(y = i,lw = 0.5, color = 'g')
    
plt.legend()
```

```python
# the pnls graphs also improve a lot
# at least we have a period of actually getting profits
data_amzn['PnL'].plot(color = 'k',lw = 1)
# green represents positive pnl
plt.plot(data_amzn[data_amzn.PnL > 0].index, data_amzn.PnL[data_amzn.PnL> 0],marker = '.',color = 'g',lw = 0,markersize = 4,label = 'positive PnL')
plt.plot(data_amzn[data_amzn.PnL < 0].index, data_amzn.PnL[data_amzn.PnL< 0],marker = '.',color = 'r',lw = 0,markersize = 4,label = 'negative PnL')
plt.legend()
```
